# Data Pre-processing

```python
def preprocess(data):
    data = data.loc[(data['points'] != None)]
    data = data.loc[(data['points'] >= 0)]
    data = data.loc[(data['points'] <= 3)]
    data['Made Basket'] = np.where(data['points'] > 0, 1, 0)
    data['Shot Attempted'] = 1
    return data
```

# PLus Minus for Lakers roster

```python
def check_players_on_court(players_on_court, condition_players):
    return all(player in players_on_court for player in condition_players)
```

```python
def check_on_court_flag(data, roster, star_players):
    players_plus_minus = {'on_court': {}, 'off_court': {}}
    for index, row in data.iterrows():
        game_id = row['game_id']
        points_scored = row['points']

        if points_scored == 0 or ('LeBron James' not in row.values and 'Anthony Davis' not in row.values):
            continue
        players_on_court = [row['a1'], row['a2'], row['a3'], row['a4'], row['a5'], row['h1'], row['h2'], row['h3'], row['h4'], row['h5']]
        check_players_flag = check_players_on_court(players_on_court, star_players)
        for player in roster:
            if player in players_on_court:
                key = 'on_court' if check_players_flag else 'off_court'
                if (game_id, player) not in players_plus_minus[key]:
                    players_plus_minus[key][(game_id, player)] = points_scored if player in players_on_court else -points_scored
                else:
                    players_plus_minus[key][(game_id, player)] += points_scored if player in players_on_court else -points_scored
    return players_plus_minus
```

```python
players_plus_minus_19 = check_on_court_flag(data_19, roster_19, star_players_19)
players_plus_minus_18 = check_on_court_flag(data_18, roster_18, star_players_18)
```

```python
def get_pm_scores(players_plus_minus_dict):
    on_court_pm_scores = {}
    off_court_pm_scores = {}
    keys = players_plus_minus_dict.keys()
    for _key in keys:
        if _key == "on_court":
            for k,v in players_plus_minus_dict[_key].items():
                if k[1] not in on_court_pm_scores:
                    on_court_pm_scores[k[1]] = [v]
                else:
                    on_court_pm_scores[k[1]].append(v)
            for k,v in on_court_pm_scores.items():
                on_court_pm_scores[k] = sum(on_court_pm_scores[k])/len(on_court_pm_scores[k])
        else:
            for k,v in players_plus_minus_dict[_key].items():
                if k[1] not in off_court_pm_scores:
                    off_court_pm_scores[k[1]] = [v]
                else:
                    off_court_pm_scores[k[1]].append(v)
            for k,v in off_court_pm_scores.items():
                off_court_pm_scores[k] = sum(off_court_pm_scores[k])/len(off_court_pm_scores[k])
    return on_court_pm_scores, off_court_pm_scores
```

```python
on_court_pm_scores_19, off_court_pm_scores_19 = get_pm_scores(players_plus_minus_19)
on_court_pm_scores_18, off_court_pm_scores_18 = get_pm_scores(players_plus_minus_18)
```

# Get Top 10 most efficient players of the league 2018-19 season

```python
#avg assist

data = pd.read_csv("/content/drive/MyDrive/Datasets/NBA Data/2018-2019 NBA Play-By-Play Data.csv", low_memory = False)


avg_assist = data.groupby(['assist', 'game_id']).size().groupby('assist').mean()

avg_assist_df = pd.DataFrame({
    'player': avg_assist.index,
    'APG': avg_assist.values
})

avg_assist_df.sort_values(by = 'APG', ascending = False)

#play length
data['play_length'] = pd.to_timedelta(data['play_length'], errors='coerce')

data = data.dropna(subset=['play_length'])

average_play_length_per_game = data.groupby(['assist'])['play_length'].mean().dt.total_seconds()

average_play_length_per_game = average_play_length_per_game.reset_index()

avg_play = pd.merge(avg_assist_df, average_play_length_per_game, left_on='player', right_on='assist', how='left')
avg_play= avg_play.drop(columns=['assist'])

avg_play.columns = ['player', 'APG', 'PlayLength(s)']


#bad plays or turnovers =
bad_play_reasons = ['bad pass', 'lost ball', 'traveling', 'turnover', 'double dribble turnover', 'palming turnover', 'illegal assist', 'turnover: shot clock', 'lane violation']

bad_play_data = data[data['reason'].str.lower().isin(bad_play_reasons)]
bad_play_totals = bad_play_data.groupby('player')['reason'].count().reset_index()

bad_play_totals.columns = ['player', 'BadPlays/TO']

playmake_df = pd.merge(avg_play, bad_play_totals, on='player', how='inner')


playmake_df = playmake_df.sort_values(by='APG', ascending=False)

playmake_df['Efficiency'] = 1 / (playmake_df['PlayLength(s)'] * (1 + playmake_df['BadPlays/TO']))

efficient_players = playmake_df.sort_values(by='Efficiency', ascending=False)
efficient_players[['player', 'APG', 'PlayLength(s)', 'BadPlays/TO', 'Efficiency']]

## JUST CHANGE THIS FOR RESPECTIVE YEAR

nba_players = ['Lonzo Ball', 'Michael Beasley', 'Reggie Bullock', 'Kentavious Caldwell-Pope', 'Alex Caruso', 'Josh Hart','Brandon Ingram','Kyle Kuzma','Rajon Rondo']


nba_players_processed = [player.lower().strip() for player in nba_players]

efficient_players_lakers_18 = efficient_players[efficient_players['player'].str.lower().isin(nba_players_processed)]
efficient_players_nba_18  = efficient_players[~efficient_players['player'].str.lower().isin(nba_players_processed)]


efficient_players_lakers_18 = efficient_players_lakers_18[efficient_players_lakers_18['APG'] > 1]
efficient_players_lakers_18 = efficient_players_lakers_18[['player', 'APG', 'PlayLength(s)', 'BadPlays/TO', 'Efficiency']].sort_values(by = 'Efficiency', ascending = False)
efficient_players_nba_18 = efficient_players_nba_18[efficient_players_nba_18['APG'] > 1]
efficient_players_nba_18 = efficient_players_nba_18[['player', 'APG', 'PlayLength(s)', 'BadPlays/TO', 'Efficiency']].sort_values(by = 'Efficiency', ascending = False)
```

```python
efficient_players_lakers_18
```

```python
efficient_players_nba_18.head(10)
```

```python
from sklearn.cluster import KMeans
lakers_eff_18 = list(efficient_players_lakers_18['player'])
nba_eff_18 = list(efficient_players_nba_18.head(10)['player'])
all_players = lakers_eff_18 + nba_eff_18
clustering_data_18 = efficient_players_lakers_18[['APG', 'PlayLength(s)', 'Efficiency']].to_numpy()
clustering_data_18 = np.concatenate((clustering_data_18, efficient_players_nba_18.head(10)[[ 'APG', 'PlayLength(s)', 'Efficiency']].to_numpy()))
```

```python
all_players
```

```python
k_means = KMeans(n_clusters=4, max_iter=20)
k_means.fit(clustering_data_18)
```

```python
k_means.labels_
```

```python
def create_clusters(lst, player_list):
    clusters = {}
    for i, value in enumerate(lst):
        if value in clusters:
            clusters[value].append(player_list[i])
        else:
            clusters[value] = [player_list[i]]
    return list(clusters.values())
```

```python
clusters_18 = create_clusters(k_means.labels_, all_players)
clusters_18
```
